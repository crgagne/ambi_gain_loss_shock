```python
def load_ambi_file():
	newlist=[]
	for ii in range(4):
		fname = 'parameters/Parameters_block'+str(ii+1)+'_order1.txt'
		with open(fname) as f:
		    content = f.readlines()
		if ii==0:
		    content=content[0].split('\r')

		for i in content:
		    newlist.append(i.replace('{','').replace('}','').split(','))
	mag_l=[]
	mag_r=[]
	p_l=[]
	p_r=[]
	r_l=[]
	r_r=[]
	for iii in newlist:
		mag_l.append(iii[0])
		mag_r.append(iii[1])
		p_l.append(iii[2])
		p_r.append(iii[3])
		r_r.append(iii[4]) # supposed to be right
		r_l.append(iii[5])
	return([mag_l,mag_r,p_l,p_r,r_r,r_l])
```

### Shuffle Blocks of 5 



```python
def count_repeats(gain_or_loss_per_block):
    morethan3 = 0
    morethan4 = 0
    morethan2 = 0 
    count = 0
    for ii,i in enumerate(gain_or_loss_per_block[1:]):
        if gain_or_loss_per_block[ii]==gain_or_loss_per_block[ii-1]:
            count+=1
        else:
            count=0
        if count>2:
            morethan3+=1
        if count>1:
            morethan2+=1
        if count>4:
            morethan4+=1
    # could have just stored these numbers #
    return(morethan4,morethan3,morethan2)
```

```python
morethan3=100
morethan2=100
while (morethan3>0 or morethan2>4):
    
    # generate permutation 
    nblocks = len(X)/5 # 60 blocks
    permutated_blocks = np.random.permutation(np.arange(nblocks))
    which_block_in_X = np.repeat(np.arange(60),5)

    # apply permutation and count the number of gain loss repeats 
    Xnew = X[which_block_in_X==permutated_blocks[0]]
    XOlist_array_new_new = XOlist_array_new[which_block_in_X==permutated_blocks[0],:]
    
    gain_or_loss_per_block = X.loc[which_block_in_X==permutated_blocks[0],'gainloss'].as_matrix()[0]
    for block in permutated_blocks[1:]:
        Xnew = pd.concat((Xnew,X[which_block_in_X==block]))
        XOlist_array_new_new = np.vstack((XOlist_array_new_new,XOlist_array_new[which_block_in_X==block,:]))
        gain_or_loss_per_block = np.append(gain_or_loss_per_block,X.loc[which_block_in_X==block,'gainloss'].as_matrix()[0])
    
    # calculate the number of repeats 
    morethan4,morethan3,morethan2=count_repeats(gain_or_loss_per_block)
    
X = Xnew.copy()
XOlist_array_new = XOlist_array_new_new.copy()

```

### Balance Check (Categorical Ambiguity)

```python
def check_balance(X):

    prob_greater_and_ambi_l = 0
    mag_greater_and_ambi_l = 0
    EV_greater_and_ambi_l = 0
    prob_greater_and_ambi_r= 0
    mag_greater_and_ambi_r = 0
    EV_greater_and_ambi_r = 0
    for trial in range(len(X)):

        mag_greater_on_l = X['ml'].as_matrix()[trial]>X['mr'].as_matrix()[trial]
        mag_greater_on_r = X['mr'].as_matrix()[trial]>X['ml'].as_matrix()[trial]
        prob_greater_on_l = X['pl'].as_matrix()[trial]>X['pr'].as_matrix()[trial]
        prob_greater_on_r = X['pr'].as_matrix()[trial]>X['pl'].as_matrix()[trial]

        EVl = X['pl'].as_matrix()[trial]*X['ml'].as_matrix()[trial]
        EVr = X['pr'].as_matrix()[trial]*X['mr'].as_matrix()[trial]
        EV_greater_on_l = EVl>EVr
        EV_greater_on_r = EVr>EVl
        amb_on_l = X['rl'].as_matrix()[trial]!=1.0
        amb_on_r = X['rr'].as_matrix()[trial]!=1.0

        if prob_greater_on_l and amb_on_l:
            prob_greater_and_ambi_l+=1
        if mag_greater_on_l and amb_on_l:
            mag_greater_and_ambi_l+=1
        if EV_greater_on_l and amb_on_l:
            EV_greater_and_ambi_l+=1    

        if prob_greater_on_r and amb_on_r:
            prob_greater_and_ambi_r+=1
        if mag_greater_on_r and amb_on_r:
            mag_greater_and_ambi_r+=1
        if EV_greater_on_r and amb_on_r:
            EV_greater_and_ambi_r+=1   

    print('total ambiguous trials: {0}'.format(trial))
    print('mag greater on left and ambiguous: {0} trials'.format(mag_greater_and_ambi_l))
    print('prob greater on left and ambiguous: {0} trials'.format(prob_greater_and_ambi_l))
    print('EV greater on left and ambiguous: {0} trials'.format(EV_greater_and_ambi_l))

    print('mag greater on right and ambiguous: {0} trials'.format(mag_greater_and_ambi_r))
    print('prob greater on right and ambiguous: {0} trials'.format(prob_greater_and_ambi_r))
    print('EV greater on right and ambiguous: {0} trials'.format(EV_greater_and_ambi_r))
```

# No Brainers

- right now it is being calculated off of pr and pl, which is not taking into account what they can see (these are the true probabilities). 


```python


#

# gain trials 
choose_right = (X['prob_o_r']>X['prob_o_l'])&(X['mr']>X['ml']) # choose right either gain or loss
choose_left = (X['prob_o_r']<X['prob_o_l'])&(X['mr']<X['ml'])
X.loc[X['gainloss']==1,'nobrainer'] = np.logical_or(choose_left,choose_right)[X['gainloss']==1]

# loss 
choose_left= (X['prob_o_r']>X['prob_o_l'])&(X['mr']<X['ml']) # choose right either gain or loss
choose_right = (X['prob_o_l']>X['prob_o_r'])&(X['ml']<X['mr'])
X.loc[X['gainloss']==2,'nobrainer'] = np.logical_or(choose_left,choose_right)[X['gainloss']==2]


X['nobrainer'] = X['nobrainer'].astype('int')
X.head()
```
