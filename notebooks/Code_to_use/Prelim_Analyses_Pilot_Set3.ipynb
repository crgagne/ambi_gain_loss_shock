```python
participant_table = pd.read_csv('../data/participant_table_pilotset3.csv',index_col=0)
participant_table.head()
```

```python
MIDs =participant_table.MID.as_matrix()
MIDs
```

- dropping subject that had missing data

```python
MIDs = MIDs[MIDs!='A3JTDNRJYVCGQD']
```

```python

for i,MID in enumerate(MIDs):
    
    # get raw data
    df=pd.read_csv('../data/part1_trial_data/'+MID+'_gainloss_raw.csv')
    
    instr_reading = instruction_reading(df)
    
    # get just trial data
    df=pd.read_csv('../data/part1_trial_data/'+MID+'_gainloss_just_trials.csv')
    
    HID = participant_table.loc[participant_table.MID==MID,'HID'].as_matrix()[0]
    # preprocess 
    df = preprocess_gainloss(df)
    df = preprocess(df)
    
    # no brainer perf
    df = better_choice_gainloss(df)
    df = right_choice(df)
    df = nobrainers(df)
    nb_perf = nobrainers_perf(df)
    
    # average reaction time 
    avg_rt = df['reaction_time'].mean()
    
    # read second data frame in 
    df=pd.read_csv('../data/part2_trial_data/'+MID+'_gainloss_just_trials.csv')
    
    # preprocess
    df=preprocess_est(df)
    
    # accuracy 
    c,p = ratio_estimation(df)
    
    avg_rt2 = df['reaction_time'].mean()
    noresp2 = calc_noresp_part2(df)
    
    tmp = pd.Series([MID,HID,instr_reading,nb_perf,avg_rt,avg_rt2,c,p,noresp2],index=['MID','HID','instr_read','nb_perf','avg_rt_part1','avg_rt_part2','est_corr','est_p','noresp2'])
    tmp = tmp.to_frame().T
    if i==0:
        summary_df = tmp
    else:
        summary_df = pd.concat([summary_df,tmp])
```

```python
resultnames = ['bic','aic','pseudoR2','pred_acc','llr_pvalue']
model_param_df = np.array(['','','',3.0])
for MID in MIDs:
    #print(MID)
    # get trial data 
    df=pd.read_csv('../data/part1_trial_data/'+MID+'_gainloss_just_trials.csv')
    df = alter_df_for_singular_subs(df)
    df = preprocess_gainloss(df)
    df = preprocess(df)

    for split in ['ambig_gain','ambig_loss','unambig_gain','unambig_loss']:
        #print(split)
        if 'unambig' in split:
            paramnum='0'
        else:
            paramnum='1'
        out = fit_model_split_amb_unamb_gain_loss(df,
                                                  whichreturn=split,
                                                  params=param_sets_split[paramnum],
                                                  zscore=True)
        modelname = out['modelname']
        #pickle.dump(out,open(savefolder+
        #        MID+'_'+modelname+'.p', "wb" ))

        for result in resultnames:
            model_summary_df.loc[(model_summary_df.MID==MID),result+'_'+split]=out[result]

        params = out['params']
        for param in params.index:
            paramn = param.replace('_loss','')
            paramn = paramn.replace('_gain','')
            paramn = paramn.replace('_amb','')
            paramn = paramn.replace('_rl','')
            row = np.array([MID,paramn,split,params[param]])
            model_param_df=np.vstack((model_param_df,row))

model_param_df = pd.DataFrame(model_param_df,columns=['MID','parameter','split','beta'])
model_param_df.drop(0,inplace=True) #df.index[0]
model_param_df['beta']=model_param_df['beta'].astype('float')
```

```python
# single bar graph for now 
def plot_params(df,stripplot=False,outlier_cutoff=None):
    plt.style.use(['seaborn-white', 'seaborn-paper'])
    matplotlib.rc("font", family="Times New Roman")
    sns.set_context('talk')
    sns.set_style('white',{'figure.facecolor':'white'})
    
    
    if outlier_cutoff is not None:
        df = df[(df.beta>-1.0*outlier_cutoff)&(df.beta<outlier_cutoff)]
    
    axis = sns.barplot(x='parameter',y='beta',hue='split',data=df,ci=95,alpha=0.4)
    
    if stripplot:
        sns.stripplot(x="parameter", y="beta",hue='split', data=df,alpha=0.2,jitter=True);

    current_palette=sns.color_palette()
    fig = plt.gcf()
    fig.suptitle('Model Parameters:',fontsize=12,x=0.55)
    sns.despine(ax=axis)
    axis.set_ylabel('beta (Prob Choose Right (except on Ambig))',fontsize=12)
    axis.set_xlabel('parameter',fontsize=12)
    axis.set_xticklabels(df.parameter.unique(),rotation=45,fontsize=12,ha='right')
    axis = plt.gca()
    
    # change name if needed
    xlabels = axis.get_xticklabels()

    fig.suptitle('')
    axis.set_title('Model Parameters (Across all Subjects)')
    axis.set_xlabel('Parameter')
    axis.set_ylabel('Group Regression Coefficients \n (Probability Choosing Ambig)')
    plt.tight_layout()
    return(fig)
```

```python
# read in STAI 
q_df=pd.read_csv('../data/questionnaire_data/pilot_set_3.csv',index_col=0)
q_df.drop(['rt','start_date'],axis=1,inplace=True)
```

```python
# merged 
model_param_df_good_subs_w_stai = model_param_df_good_subs.merge(q_df,on='MID',how='left')
model_param_df_good_subs_w_stai.head()
```

```python
betas=model_param_df_good_subs_w_stai.loc[(model_param_df_good_subs_w_stai['parameter']=='intercept')&
                                 (model_param_df_good_subs_w_stai['split']=='ambig_gain'),'beta'].as_matrix()
stais=model_param_df_good_subs_w_stai.loc[(model_param_df_good_subs_w_stai['parameter']=='intercept')&
                                 (model_param_df_good_subs_w_stai['split']=='ambig_gain'),'stai_trait'].as_matrix()
```

```python
from scipy.stats import pearsonr
```

```python
outlier_cutoff = 6

fig,axes = plt.subplots(2,3,figsize=(12,6))
for pi,param in enumerate(['intercept','prob_diff','sqrt_prop_revealed']):
    for si,split in enumerate(['ambig_gain','ambig_loss']):
        betas=model_param_df_good_subs_w_stai.loc[(model_param_df_good_subs_w_stai['parameter']==param)&
                                 (model_param_df_good_subs_w_stai['split']==split),'beta'].as_matrix()
        stais=model_param_df_good_subs_w_stai.loc[(model_param_df_good_subs_w_stai['parameter']==param)&
                                         (model_param_df_good_subs_w_stai['split']==split),'stai_trait'].as_matrix()
        include = np.logical_and(betas<outlier_cutoff,betas>-outlier_cutoff)
        excluded = np.sum(include==False)
        betas = betas[include]
        stais = stais[include]
        axes[si,pi].scatter(stais,betas)
        c,p = pearsonr(stais,betas)
        axes[si,pi].set_title(param+' '+split+' \ncorr:'+str(np.round(c,3))+' p:'+str(np.round(p,3))+' exc:'+str(excluded))
        sns.despine()
plt.tight_layout()
```
