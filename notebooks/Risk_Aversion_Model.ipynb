Risk Aversion is used to described the trade-off between probability and magnitude. 
Models using risk aversion usually model the choice between two options as an expected value calculation, where the magnitude is raised to a power. 

$$ev = p_1m_1^p -p_2m_2^p $$

If the $p<1$, higher magnitudes have a proportionally smaller effect than probability (ie. risk aversion). 
If $p>1$, behavior is characterized as risk seeking.
